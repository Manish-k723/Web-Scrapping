This approach is making scarpping harder because for data security ambition-box have added multiple entities into same tag which is making it harder for extraction because of no differentiation. So, instead of doing this. We will firstly take single company at a time and work on it and then loop over all. So, let's do it.

```python
company = soup.find_all('div', class_ = 'company-content-wrapper')
```

```python
company[0].find('p',class_ = 'description body-small').text.strip()
```

```python
title = []
rating = []
review_count = []
ctype = []
hq = []
old = []
employee = []
des = []
for i in company:
  title.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_ = 'rating').text.strip())
  review_count.append(i.find('a',class_ = 'review-count').text.strip())
  ctype.append(i.find_all('p',class_ = 'infoEntity')[0].text.strip())
  hq.append(i.find_all('p',class_ = 'infoEntity')[1].text.strip())
  old.append(i.find_all('p',class_ = 'infoEntity')[2].text.strip())
  employee.append(i.find_all('p',class_ = 'infoEntity')[3].text.strip())
  des.append(i.find('p',class_ = 'description body-small').text.strip())

d = {'Company_name': title, 'Overall-rating':rating,'Total Reviews':review_count,'Type of Company': ctype,'Headquarter': hq,'How Old': old,
     'Total employees': employee,'Description': des}
df = pd.DataFrame(d)
df
```

Ohh, this is data of only single page, Let's create a final dataframe for 100 pages

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
import warnings
```

```python
finall_df = pd.DataFrame()
warnings.filterwarnings('ignore')
for j in range(1,3):
  url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)

  webpage = requests.get(url,headers = headers).text

  soup = BeautifulSoup(webpage,'lxml')
  company = soup.find_all('div',class_ = 'company-content-wrapper')

  title = []
  rating = []
  review_count = []
  ctype = []
  hq = []
  old = []
  employee = []
  des = []
  for i in company:
    title.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_ = 'rating').text.strip())
    review_count.append(i.find('a',class_ = 'review-count').text.strip())
    try:
      ctype.append(i.find_all('p',class_ = 'infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:
      hq.append(i.find_all('p',class_ = 'infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    try:
      old.append(i.find_all('p',class_ = 'infoEntity')[2].text.strip())
    except:
      old.append(np.nan)
    try:
      employee.append(i.find_all('p',class_ = 'infoEntity')[3].text.strip())
    except:
      employee.append(np.nan)
    try:
      des.append(i.find('p',class_ = 'description body-small').text.strip())
    except:
      des.append(np.nan)

  d = {'Company_name': title, 'Overall-rating':rating,'Total Reviews':review_count,'Type of Company': ctype,'Headquarter': hq,'How Old': old,
     'Total employees': employee,'Description': des}
  df = pd.DataFrame(d)
  finall_df = finall_df.append(df)
```
